# DBSCAN and Clustering Review

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200,
                  noise=0.05,
                  random_state=0)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

## 1. K-means clustering

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2,random_state=0)
y_km = km.fit_predict(X)

plt.scatter(X[y_km==0,0],X[y_km==0,1],
            c='lightblue',
            marker='o', 
            s=40, 
            label='cluster 1')
plt.scatter(X[y_km==1,0],X[y_km==1,1],
            c='red',
            marker='s',
            s=40,
            label='cluster 2'
           )
plt.title('K-means Clustering')
```

## 2. Agglomerative clustering 

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=2,
                             affinity='euclidean',
                             linkage='complete')
y_ac = ac.fit_predict(X)

plt.scatter(X[y_ac==0,0],X[y_ac==0,1],
            c='lightblue',
            marker='o', 
            s=40, 
            label='cluster 1')
plt.scatter(X[y_ac==1,0],X[y_ac==1,1],
            c='red',
            marker='s',
            s=40,
            label='cluster 2'
           )
plt.title('Agglomerative Clustering')
```

## 3. DBSCAN 

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.25,
            min_samples=4,
           metric='euclidean')

y_db = db.fit_predict(X)

plt.scatter(X[y_db==0,0],X[y_db==0,1],
            c='lightblue',
            marker='o', 
            s=40, 
            label='cluster 1')
plt.scatter(X[y_db==1,0],X[y_db==1,1],
            c='red',
            marker='s',
            s=40,
            label='cluster 2'
           )
plt.title('DBSCAN')
```

```python
db2 = GridSearchCV(estimator = DBSCAN(eps=0.2, min_samples=5, metric='euclidean'),\
                   param_grid={'eps':Eps,'min_samples':Min_samples},\
                   scoring='adjusted_mutual_info_score',\
                   cv = None,\
                   n_jobs=-1)
```

```python
db2.fit_predict(X)
```

```python
eps = []
min_samples = []
scores = []
for i in Eps:
    for j in Min_samples:
        db = DBSCAN(eps=i, min_samples = j, metric='euclidean')
        y_db=db.fit_predict(X)
        if (len(set(y_db)) > 1):
            score = metrics.silhouette_score(X, db.labels_, metric='euclidean')
            eps.append(i)
            min_samples.append(j)
            scores.append(score)
            
```
